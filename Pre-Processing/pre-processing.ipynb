```python
from PIL import Image
import os
import shutil
```

```python
def prepare_dirs():
    os.chdir('./Data')
    if not os.path.exists('Pre-Processed'):
        os.mkdir('Pre-Processed')
    os.chdir('./Pre-Processed')
    if not os.path.exists('Covid'):
        os.mkdir('Covid')
    if not os.path.exists('Normal'):
        os.mkdir('Normal')
    #if not os.path.exists('Resized_Images'):
    #    os.mkdir('Resized_Images')
    #os.chdir('./Resized_Images')
    #if not os.path.exists('Covid-19'):
    #    os.mkdir('Covid-19')
    #if not os.path.exists('Normal'):
    #    os.mkdir('Normal')
    #os.chdir('..')
    os.chdir('..')
```

```python
def resize_and_rename_image(file, directory, label, name, count):
    size = (224,224)
    if (file.endswith('.jpg') or file.endswith('.jpeg') or file.endswith('.png')):
        image_name = '{}_{}'.format(name,count)
        image_path = directory+'/'+file
        i = Image.open(image_path)
        fsplits = file.split('.')
        fext = fsplits[-1]
        img = i.resize(size)
        img = img.convert('L')
        img.save('./Pre-Processed/{}/{}.{}'.format(label,image_name,fext))
```

```python
def process_images():
    covid_name = 'Covid'
    normal_name = 'Normal'
    covid_count = 0
    normal_count = 0
    
    for dataset in os.listdir('.'):
        if(dataset == '3-classes'):
            for folder in os.listdir('./{}'.format(dataset)):
                for sub_folder in os.listdir('./{}/{}'.format(dataset,folder)):
                    directory = './{}/{}/{}'.format(dataset,folder,sub_folder)
                    if(sub_folder == 'Covid-19'):
                        for file in os.listdir('./{}/{}/{}'.format(dataset,folder,sub_folder)):
                            covid_count += 1
                            resize_and_rename_image(file, directory, 'Covid', covid_name, covid_count)    
                        #source_dir = "./{}/{}/{}".format(dataset,folder,sub_folder)
                        #destination_dir = "./Pre-Processed/Covid-19"
                        #shutil.copytree(source_dir, destination_dir,dirs_exist_ok=True)
                    elif(sub_folder == 'Normal'):
                        for file in os.listdir('./{}/{}/{}'.format(dataset,folder,sub_folder)):
                            normal_count += 1
                            resize_and_rename_image(file, directory, sub_folder, normal_name, normal_count)
                    else:
                        continue

        elif(dataset == 'COVID-19_Radiography_Dataset'):
            for folder in os.listdir('./{}'.format(dataset)):
                directory = './{}/{}'.format(dataset,folder)
                if(folder == 'COVID'):
                    for file in os.listdir('./{}/{}'.format(dataset,folder)):
                        covid_count += 1
                        resize_and_rename_image(file, directory, 'Covid', covid_name, covid_count) 
                    #source_dir = "./{}/{}".format(dataset,folder)
                    #destination_dir = "./Covid-19"
                    #shutil.copytree(source_dir, destination_dir,dirs_exist_ok=True)
                elif(folder == 'Normal'):
                    for file in os.listdir('./{}/{}'.format(dataset,folder)):
                        normal_count += 1
                        resize_and_rename_image(file, directory, folder, normal_name, normal_count) 
                else:
                    continue
```

```python
prepare_dirs()
process_images()
```
